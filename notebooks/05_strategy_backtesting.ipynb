```python


# #!/usr/bin/env python3
# """
# Investment Strategy Backtesting for FinDocGPT

# This script demonstrates comprehensive backtesting of investment strategies including:
# - Technical analysis strategies
# - Sentiment-based strategies  
# - Portfolio optimization
# - Risk management and performance metrics

# Author: FinDocGPT Team
# Date: 2025
# """

# import pandas as pd
# import numpy as np
# import matplotlib.pyplot as plt
# import seaborn as sns
# from datetime import datetime, timedelta
# import warnings
# warnings.filterwarnings('ignore')

# # Financial analysis
# import yfinance as yf
# from scipy import stats
# from scipy.optimize import minimize

# # Portfolio optimization
# try:
#     from pypfopt import EfficientFrontier, risk_models, expected_returns
#     pypfopt_available = True
# except ImportError:
#     pypfopt_available = False
#     print("PyPortfolioOpt not available - using custom implementation")

# # Additional imports for report generation
# from pathlib import Path
# import json


# class MarketDataLoader:
#     """Class to handle market data loading and preparation"""
    
#     def __init__(self):
#         self.symbols = {
#             'Tech': ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'TSLA', 'NVDA'],
#             'Finance': ['JPM', 'BAC', 'WFC', 'GS', 'C'],
#             'Healthcare': ['JNJ', 'PFE', 'UNH', 'ABBV'],
#             'Energy': ['XOM', 'CVX', 'COP'],
#             'ETFs': ['SPY', 'QQQ', 'IWM']
#         }
#         self.all_symbols = [symbol for sector_symbols in self.symbols.values() 
#                            for symbol in sector_symbols]
#         self.market_data = {}
#         self.price_data = pd.DataFrame()
#         self.successful_symbols = []
#         self.failed_symbols = []
    
#     def load_data(self, period='3y'):
#         """Load market data for all symbols"""
#         print(f"üåç Investment Universe: {len(self.all_symbols)} symbols across {len(self.symbols)} sectors")
#         print(f"üìÖ Analysis period: {period}")
        
#         print("\nüìä Downloading market data...")
        
#         for symbol in self.all_symbols:
#             try:
#                 data = yf.download(symbol, period=period, progress=False)
#                 if not data.empty and len(data) > 100:  # Minimum data requirement
#                     self.market_data[symbol] = data
#                     self.successful_symbols.append(symbol)
#                     print(f"‚úÖ {symbol}: {len(data)} days")
#                 else:
#                     self.failed_symbols.append(symbol)
#                     print(f"‚ùå {symbol}: Insufficient data")
#             except Exception as e:
#                 self.failed_symbols.append(symbol)
#                 print(f"‚ùå {symbol}: Error - {e}")
        
#         print(f"\n‚úÖ Successfully loaded data for {len(self.successful_symbols)} symbols")
#         if self.failed_symbols:
#             print(f"‚ùå Failed to load: {self.failed_symbols}")
        
#         # Create combined price DataFrame
#         for symbol, data in self.market_data.items():
#             self.price_data[symbol] = data['Close']
        
#         self.price_data = self.price_data.dropna()
#         print(f"\nüìä Combined dataset: {self.price_data.shape}")
#         print(f"üìÖ Date range: {self.price_data.index.min().date()} to {self.price_data.index.max().date()}")
        
#         return self.price_data


# class TechnicalStrategies:
#     """Implementation of various technical analysis strategies"""
    
#     def __init__(self, price_data):
#         self.price_data = price_data
#         self.returns = price_data.pct_change().dropna()
    
#     def moving_average_crossover(self, short_window=20, long_window=50):
#         """Moving average crossover strategy"""
#         signals = pd.DataFrame(index=self.price_data.index)
        
#         for symbol in self.price_data.columns:
#             prices = self.price_data[symbol]
            
#             # Calculate moving averages
#             short_ma = prices.rolling(window=short_window).mean()
#             long_ma = prices.rolling(window=long_window).mean()
            
#             # Generate signals
#             signal = pd.Series(0, index=prices.index)
#             signal[short_ma > long_ma] = 1  # Buy signal
#             signal[short_ma < long_ma] = -1  # Sell signal
            
#             signals[symbol] = signal
        
#         return signals.fillna(0)
    
#     def rsi_strategy(self, window=14, oversold=30, overbought=70):
#         """RSI-based mean reversion strategy"""
#         signals = pd.DataFrame(index=self.price_data.index)
        
#         for symbol in self.price_data.columns:
#             prices = self.price_data[symbol]
            
#             # Calculate RSI
#             delta = prices.diff()
#             gain = (delta.where(delta > 0, 0)).rolling(window=window).mean()
#             loss = (-delta.where(delta < 0, 0)).rolling(window=window).mean()
#             rs = gain / loss
#             rsi = 100 - (100 / (1 + rs))
            
#             # Generate signals
#             signal = pd.Series(0, index=prices.index)
#             signal[rsi < oversold] = 1  # Buy when oversold
#             signal[rsi > overbought] = -1  # Sell when overbought
            
#             signals[symbol] = signal
        
#         return signals.fillna(0)
    
#     def bollinger_bands_strategy(self, window=20, num_std=2):
#         """Bollinger Bands mean reversion strategy"""
#         signals = pd.DataFrame(index=self.price_data.index)
        
#         for symbol in self.price_data.columns:
#             prices = self.price_data[symbol]
            
#             # Calculate Bollinger Bands
#             sma = prices.rolling(window=window).mean()
#             std = prices.rolling(window=window).std()
#             upper_band = sma + (std * num_std)
#             lower_band = sma - (std * num_std)
            
#             # Generate signals
#             signal = pd.Series(0, index=prices.index)
#             signal[prices < lower_band] = 1  # Buy when below lower band
#             signal[prices > upper_band] = -1  # Sell when above upper band
            
#             signals[symbol] = signal
        
#         return signals.fillna(0)
    
#     def momentum_strategy(self, lookback=20, hold_period=5):
#         """Momentum strategy based on recent returns"""
#         signals = pd.DataFrame(index=self.price_data.index)
        
#         for symbol in self.price_data.columns:
#             prices = self.price_data[symbol]
            
#             # Calculate momentum (return over lookback period)
#             momentum = prices.pct_change(periods=lookback)
            
#             # Generate signals based on momentum percentiles
#             signal = pd.Series(0, index=prices.index)
            
#             # Buy top 30% momentum, sell bottom 30%
#             for date in momentum.index:
#                 if pd.notna(momentum[date]):
#                     rolling_quantile_high = momentum.rolling(window=60).quantile(0.7)[date]
#                     rolling_quantile_low = momentum.rolling(window=60).quantile(0.3)[date]
                    
#                     if pd.notna(rolling_quantile_high) and momentum[date] > rolling_quantile_high:
#                         signal[date] = 1
#                     elif pd.notna(rolling_quantile_low) and momentum[date] < rolling_quantile_low:
#                         signal[date] = -1
            
#             signals[symbol] = signal
        
#         return signals.fillna(0)


# class StrategyBacktester:
#     """Comprehensive backtesting engine"""
    
#     def __init__(self, price_data, initial_capital=100000, commission=0.001):
#         self.price_data = price_data
#         self.returns = price_data.pct_change().dropna()
#         self.initial_capital = initial_capital
#         self.commission = commission
    
#     def backtest_strategy(self, signals, strategy_name):
#         """Backtest a single strategy"""
        
#         # Initialize portfolio
#         portfolio_value = pd.Series(self.initial_capital, index=self.returns.index)
#         positions = pd.DataFrame(0, index=self.returns.index, columns=self.returns.columns)
#         trades = []
#         cash = self.initial_capital
        
#         # Align signals with returns
#         aligned_signals = signals.reindex(self.returns.index).fillna(0)
        
#         for i, date in enumerate(self.returns.index):
#             if i == 0:
#                 continue
            
#             prev_date = self.returns.index[i-1]
#             current_positions = positions.loc[prev_date].copy()
            
#             # Update portfolio value based on returns
#             if i > 1:
#                 position_values = current_positions * self.price_data.loc[date]
#                 portfolio_value[date] = cash + position_values.sum()
            
#             # Process signals for each asset
#             for symbol in self.returns.columns:
#                 signal = aligned_signals.loc[date, symbol]
#                 current_pos = current_positions[symbol]
#                 price = self.price_data.loc[date, symbol]
                
#                 if pd.isna(price) or pd.isna(signal):
#                     continue
                
#                 # Calculate position size (equal weight among active positions)
#                 active_signals = aligned_signals.loc[date]
#                 num_active = (active_signals == 1).sum()
                
#                 if signal == 1 and current_pos == 0 and num_active > 0:  # Buy signal
#                     position_value = portfolio_value[date] / num_active * 0.95  # 95% of available capital
#                     shares = position_value / price
#                     cost = shares * price * (1 + self.commission)
                    
#                     if cost <= cash:
#                         current_positions[symbol] = shares
#                         cash -= cost
#                         trades.append({
#                             'date': date,
#                             'symbol': symbol,
#                             'action': 'BUY',
#                             'shares': shares,
#                             'price': price,
#                             'value': cost
#                         })
                
#                 elif signal == -1 and current_pos > 0:  # Sell signal
#                     proceeds = current_pos * price * (1 - self.commission)
#                     cash += proceeds
#                     trades.append({
#                         'date': date,
#                         'symbol': symbol,
#                         'action': 'SELL',
#                         'shares': current_pos,
#                         'price': price,
#                         'value': proceeds
#                     })
#                     current_positions[symbol] = 0
            
#             positions.loc[date] = current_positions
        
#         # Calculate final portfolio value
#         final_positions = positions.iloc[-1]
#         final_prices = self.price_data.iloc[-1]
#         final_portfolio_value = cash + (final_positions * final_prices).sum()
#         portfolio_value.iloc[-1] = final_portfolio_value
        
#         # Calculate strategy returns
#         strategy_returns = portfolio_value.pct_change().dropna()
        
#         return {
#             'portfolio_value': portfolio_value,
#             'strategy_returns': strategy_returns,
#             'positions': positions,
#             'trades': trades,
#             'final_value': final_portfolio_value
#         }
    
#     def calculate_performance_metrics(self, strategy_returns, benchmark_returns=None):
#         """Calculate comprehensive performance metrics"""
        
#         # Basic metrics
#         total_return = (strategy_returns + 1).prod() - 1
#         annual_return = (1 + total_return) ** (252 / len(strategy_returns)) - 1
#         volatility = strategy_returns.std() * np.sqrt(252)
        
#         # Risk metrics
#         sharpe_ratio = annual_return / volatility if volatility > 0 else 0
        
#         # Drawdown analysis
#         cumulative_returns = (1 + strategy_returns).cumprod()
#         peak = cumulative_returns.expanding().max()
#         drawdown = (cumulative_returns - peak) / peak
#         max_drawdown = drawdown.min()
        
#         # Additional metrics
#         positive_days = (strategy_returns > 0).sum()
#         total_days = len(strategy_returns)
#         win_rate = positive_days / total_days
        
#         metrics = {
#             'total_return': total_return,
#             'annual_return': annual_return,
#             'volatility': volatility,
#             'sharpe_ratio': sharpe_ratio,
#             'max_drawdown': abs(max_drawdown),
#             'win_rate': win_rate,
#             'total_trades': len(strategy_returns),
#             'calmar_ratio': annual_return / abs(max_drawdown) if max_drawdown != 0 else 0
#         }
        
#         # Benchmark comparison
#         if benchmark_returns is not None:
#             benchmark_annual = (1 + ((benchmark_returns + 1).prod() - 1)) ** (252 / len(benchmark_returns)) - 1
#             excess_return = annual_return - benchmark_annual
            
#             # Beta calculation
#             covariance = np.cov(strategy_returns.dropna(), benchmark_returns.dropna())[0][1]
#             benchmark_variance = np.var(benchmark_returns.dropna())
#             beta = covariance / benchmark_variance if benchmark_variance > 0 else 0
            
#             metrics.update({
#                 'benchmark_annual_return': benchmark_annual,
#                 'excess_return': excess_return,
#                 'beta': beta,
#                 'alpha': annual_return - (0.02 + beta * (benchmark_annual - 0.02))  # Assuming 2% risk-free rate
#             })
        
#         return metrics


# class PortfolioOptimizer:
#     """Portfolio optimization using Modern Portfolio Theory"""
    
#     def __init__(self, returns_data):
#         self.returns = returns_data.dropna()
#         self.mean_returns = self.returns.mean() * 252
#         self.cov_matrix = self.returns.cov() * 252
#         self.n_assets = len(self.returns.columns)
    
#     def portfolio_performance(self, weights):
#         """Calculate portfolio performance metrics"""
#         portfolio_return = np.dot(weights, self.mean_returns)
#         portfolio_volatility = np.sqrt(np.dot(weights.T, np.dot(self.cov_matrix, weights)))
#         sharpe_ratio = portfolio_return / portfolio_volatility if portfolio_volatility > 0 else 0
#         return portfolio_return, portfolio_volatility, sharpe_ratio
    
#     def optimize_portfolio(self, objective='sharpe', target_return=None):
#         """Optimize portfolio based on objective"""
        
#         # Constraints
#         constraints = [{'type': 'eq', 'fun': lambda x: np.sum(x) - 1}]  # Weights sum to 1
        
#         if target_return is not None:
#             constraints.append({
#                 'type': 'eq', 
#                 'fun': lambda x: np.dot(x, self.mean_returns) - target_return
#             })
        
#         # Bounds (no short selling)
#         bounds = tuple((0, 1) for _ in range(self.n_assets))
        
#         # Initial guess
#         x0 = np.array([1/self.n_assets] * self.n_assets)
        
#         # Objective functions
#         if objective == 'sharpe':
#             def objective_func(weights):
#                 _, vol, sharpe = self.portfolio_performance(weights)
#                 return -sharpe  # Negative because we're minimizing
#         elif objective == 'volatility':
#             def objective_func(weights):
#                 _, vol, _ = self.portfolio_performance(weights)
#                 return vol
#         else:
#             raise ValueError("Objective must be 'sharpe' or 'volatility'")
        
#         # Optimize
#         result = minimize(objective_func, x0, method='SLSQP', 
#                          bounds=bounds, constraints=constraints)
        
#         if result.success:
#             optimal_weights = result.x
#             ret, vol, sharpe = self.portfolio_performance(optimal_weights)
            
#             return {
#                 'weights': dict(zip(self.returns.columns, optimal_weights)),
#                 'expected_return': ret,
#                 'volatility': vol,
#                 'sharpe_ratio': sharpe,
#                 'success': True
#             }
#         else:
#             return {'success': False, 'message': result.message}
    
#     def efficient_frontier(self, num_points=50):
#         """Generate efficient frontier"""
#         min_ret = self.mean_returns.min()
#         max_ret = self.mean_returns.max()
#         target_returns = np.linspace(min_ret, max_ret, num_points)
        
#         efficient_portfolios = []
        
#         for target_ret in target_returns:
#             try:
#                 result = self.optimize_portfolio('volatility', target_ret)
#                 if result.get('success'):
#                     efficient_portfolios.append({
#                         'return': result['expected_return'],
#                         'volatility': result['volatility'],
#                         'sharpe': result['sharpe_ratio']
#                     })
#             except:
#                 continue
        
#         return pd.DataFrame(efficient_portfolios)


# class RiskAnalyzer:
#     """Comprehensive risk analysis"""
    
#     def __init__(self, returns_data):
#         self.returns = returns_data.dropna()
    
#     def calculate_var(self, confidence_level=0.05, portfolio_weights=None):
#         """Calculate Value at Risk"""
#         if portfolio_weights is not None:
#             portfolio_returns = (self.returns * portfolio_weights).sum(axis=1)
#         else:
#             # Equal weight portfolio
#             portfolio_returns = self.returns.mean(axis=1)
        
#         var = np.percentile(portfolio_returns, confidence_level * 100)
#         return var
    
#     def calculate_cvar(self, confidence_level=0.05, portfolio_weights=None):
#         """Calculate Conditional Value at Risk (Expected Shortfall)"""
#         if portfolio_weights is not None:
#             portfolio_returns = (self.returns * portfolio_weights).sum(axis=1)
#         else:
#             portfolio_returns = self.returns.mean(axis=1)
        
#         var = self.calculate_var(confidence_level, portfolio_weights)
#         cvar = portfolio_returns[portfolio_returns <= var].mean()
#         return cvar
    
#     def correlation_analysis(self):
#         """Analyze asset correlations"""
#         correlation_matrix = self.returns.corr()
        
#         # Find highest and lowest correlations
#         corr_pairs = []
#         for i in range(len(correlation_matrix.columns)):
#             for j in range(i+1, len(correlation_matrix.columns)):
#                 corr_pairs.append({
#                     'asset1': correlation_matrix.columns[i],
#                     'asset2': correlation_matrix.columns[j],
#                     'correlation': correlation_matrix.iloc[i, j]
#                 })
        
#         corr_df = pd.DataFrame(corr_pairs).sort_values('correlation', ascending=False)
        
#         return {
#             'correlation_matrix': correlation_matrix,
#             'highest_correlations': corr_df.head(10),
#             'lowest_correlations': corr_df.tail(10)
#         }
    
#     def stress_testing(self, scenarios):
#         """Perform stress testing on portfolios"""
#         stress_results = {}
        
#         for scenario_name, scenario_returns in scenarios.items():
#             # Apply scenario to each asset
#             stressed_returns = {}
#             for asset in self.returns.columns:
#                 if asset in scenario_returns:
#                     stressed_returns[asset] = scenario_returns[asset]
#                 else:
#                     # Default stress (e.g., -10%)
#                     stressed_returns[asset] = -0.10
            
#             stress_results[scenario_name] = stressed_returns
        
#         return stress_results


# class VisualizationManager:
#     """Handle all visualization tasks"""
    
#     @staticmethod
#     def plot_strategy_signals(price_data, strategies, sample_symbol=None):
#         """Visualize strategy signals for a sample stock"""
#         if sample_symbol is None:
#             sample_symbol = 'AAPL' if 'AAPL' in price_data.columns else price_data.columns[0]
        
#         sample_data = price_data[sample_symbol].tail(252)  # Last year
        
#         fig, axes = plt.subplots(2, 2, figsize=(15, 10))
#         axes = axes.flatten()
        
#         for i, (strategy_name, signals) in enumerate(strategies.items()):
#             ax = axes[i]
            
#             # Plot price
#             ax.plot(sample_data.index, sample_data.values, label='Price', alpha=0.7)
            
#             # Plot buy/sell signals
#             strategy_signals = signals[sample_symbol].tail(252)
#             buy_dates = strategy_signals[strategy_signals == 1].index
#             sell_dates = strategy_signals[strategy_signals == -1].index
            
#             if len(buy_dates) > 0:
#                 ax.scatter(buy_dates, sample_data[buy_dates], 
#                           color='green', marker='^', s=50, label='Buy Signal')
            
#             if len(sell_dates) > 0:
#                 ax.scatter(sell_dates, sample_data[sell_dates], 
#                           color='red', marker='v', s=50, label='Sell Signal')
            
#             ax.set_title(f'{strategy_name} - {sample_symbol}')
#             ax.legend()
#             ax.grid(True, alpha=0.3)
        
#         plt.tight_layout()
#         plt.show()
    
#     @staticmethod
#     def plot_strategy_performance(backtest_results, benchmark_returns=None):
#         """Visualize strategy performance"""
#         fig, axes = plt.subplots(2, 2, figsize=(15, 10))
        
#         # Create performance dataframe
#         performance_df = pd.DataFrame({
#             name: results['metrics'] 
#             for name, results in backtest_results.items()
#         }).T
        
#         # 1. Cumulative returns
#         ax1 = axes[0, 0]
#         for strategy_name, results in backtest_results.items():
#             portfolio_value = results['result']['portfolio_value']
#             cumulative_return = portfolio_value / portfolio_value.iloc[0]
#             ax1.plot(cumulative_return.index, cumulative_return.values, 
#                     label=strategy_name, linewidth=2)
        
#         # Add benchmark if available
#         if benchmark_returns is not None:
#             benchmark_cumulative = (1 + benchmark_returns).cumprod()
#             ax1.plot(benchmark_cumulative.index, benchmark_cumulative.values, 
#                     label='SPY Benchmark', color='black', linestyle='--', alpha=0.7)
        
#         ax1.set_title('Cumulative Returns Comparison')
#         ax1.legend()
#         ax1.grid(True, alpha=0.3)
        
#         # 2. Annual returns vs volatility
#         ax2 = axes[0, 1]
#         returns = performance_df['annual_return'].values
#         volatilities = performance_df['volatility'].values
#         sharpe_ratios = performance_df['sharpe_ratio'].values
        
#         scatter = ax2.scatter(volatilities, returns, s=100, c=sharpe_ratios, 
#                              cmap='RdYlGn', alpha=0.7)
#         ax2.set_xlabel('Volatility (Annual)')
#         ax2.set_ylabel('Annual Return')
#         ax2.set_title('Risk-Return Profile')
#         ax2.grid(True, alpha=0.3)
        
#         # Add strategy labels
#         for i, strategy in enumerate(performance_df.index):
#             ax2.annotate(strategy, (volatilities[i], returns[i]), 
#                         xytext=(5, 5), textcoords='offset points', fontsize=9)
        
#         plt.colorbar(scatter, ax=ax2, label='Sharpe Ratio')
        
#         # 3. Drawdown analysis
#         ax3 = axes[1, 0]
#         for strategy_name, results in backtest_results.items():
#             portfolio_value = results['result']['portfolio_value']
#             cumulative = portfolio_value / portfolio_value.iloc[0]
#             peak = cumulative.expanding().max()
#             drawdown = (cumulative - peak) / peak
#             ax3.fill_between(drawdown.index, drawdown.values, 0, alpha=0.3, label=strategy_name)
        
#         ax3.set_title('Strategy Drawdowns')
#         ax3.set_ylabel('Drawdown')
#         ax3.legend()
#         ax3.grid(True, alpha=0.3)
        
#         # 4. Performance metrics comparison
#         ax4 = axes[1, 1]
#         metrics_to_plot = ['annual_return', 'sharpe_ratio', 'max_drawdown']
#         x_pos = np.arange(len(performance_df.index))
#         width = 0.25
        
#         for i, metric in enumerate(metrics_to_plot):
#             values = performance_df[metric].values
#             if metric == 'max_drawdown':
#                 values = -values  # Make negative for better visualization
#             ax4.bar(x_pos + i*width, values, width, label=metric.replace('_', ' ').title())
        
#         ax4.set_xlabel('Strategy')
#         ax4.set_ylabel('Value')
#         ax4.set_title('Key Performance Metrics')
#         ax4.set_xticks(x_pos + width)
#         ax4.set_xticklabels(performance_df.index, rotation=45)
#         ax4.legend()
#         ax4.grid(True, alpha=0.3)
        
#         plt.tight_layout()
#         plt.show()
    
#     @staticmethod
#     def plot_portfolio_optimization(optimizer, max_sharpe_portfolio, min_vol_portfolio, efficient_frontier):
#         """Visualize portfolio optimization results"""
#         fig, axes = plt.subplots(1, 2, figsize=(15, 6))
        
#         # 1. Efficient frontier
#         ax1 = axes[0]
#         if not efficient_frontier.empty:
#             ax1.plot(efficient_frontier['volatility'], efficient_frontier['return'], 
#                     'b-', linewidth=2, label='Efficient Frontier')
        
#         # Plot optimal portfolios
#         if max_sharpe_portfolio['success']:
#             ax1.scatter(max_sharpe_portfolio['volatility'], max_sharpe_portfolio['expected_return'],
#                        color='red', s=100, label='Max Sharpe', marker='*', zorder=5)
        
#         if min_vol_portfolio['success']:
#             ax1.scatter(min_vol_portfolio['volatility'], min_vol_portfolio['expected_return'],
#                        color='green', s=100, label='Min Volatility', marker='o', zorder=5)
        
#         # Plot individual assets
#         for asset in optimizer.returns.columns[:10]:  # Show first 10 assets
#             ret = optimizer.mean_returns[asset]
#             vol = np.sqrt(optimizer.cov_matrix.loc[asset, asset])
#             ax1.scatter(vol, ret, alpha=0.5, s=30)
#             ax1.annotate(asset, (vol, ret), xytext=(5, 5), 
#                         textcoords='offset points', fontsize=8)
        
#         ax1.set_xlabel('Volatility (Risk)')
#         ax1.set_ylabel('Expected Return')
#         ax1.set_title('Efficient Frontier')
#         ax1.legend()
#         ax1.grid(True, alpha=0.3)
        
#         # 2. Portfolio weights comparison
#         ax2 = axes[1]
        
#         if max_sharpe_portfolio['success'] and min_vol_portfolio['success']:
#             # Compare top holdings
#             sharpe_weights = pd.Series(max_sharpe_portfolio['weights']).sort_values(ascending=False)
#             min_vol_weights = pd.Series(min_vol_portfolio['weights']).sort_values(ascending=False)
            
#             max_sharpe_top = sharpe_weights.head(8)
#             min_vol_top = min_vol_weights.head(8)
            
#             # Get common assets
#             common_assets = list(set(max_sharpe_top.index) | set(min_vol_top.index))
            
#             x_pos = np.arange(len(common_assets))
#             width = 0.35
            
#             sharpe_values = [max_sharpe_top.get(asset, 0) for asset in common_assets]
#             minvol_values = [min_vol_top.get(asset, 0) for asset in common_assets]
            
#             ax2.bar(x_pos - width/2, sharpe_values, width, label='Max Sharpe', alpha=0.7)
#             ax2.bar(x_pos + width/2, minvol_values, width, label='Min Volatility', alpha=0.7)
            
#             ax2.set_xlabel('Assets')
#             ax2.set_ylabel('Portfolio Weight')
#             ax2.set_title('Portfolio Weights Comparison')
#             ax2.set_xticks(x_pos)
#             ax2.set_xticklabels(common_assets, rotation=45)
#             ax2.legend()
#             ax2.grid(True, alpha=0.3)
        
#         plt.tight_layout()
#         plt.show()
    
#     @staticmethod
#     def plot_risk_analysis(risk_analyzer, max_sharpe_portfolio, min_vol_portfolio, 
#                           sharpe_var, sharpe_cvar, minvol_var, minvol_cvar, backtester):
#         """Visualize risk analysis results"""
#         fig, axes = plt.subplots(2, 2, figsize=(15, 10))
        
#         corr_analysis = risk_analyzer.correlation_analysis()
        
#         # 1. Correlation heatmap
#         ax1 = axes[0, 0]
#         corr_matrix = corr_analysis['correlation_matrix']
#         # Show subset for readability
#         subset_corr = corr_matrix.iloc[:15, :15] if len(corr_matrix) > 15 else corr_matrix
        
#         sns.heatmap(subset_corr, annot=False, cmap='RdYlBu_r', center=0, 
#                    square=True, ax=ax1, cbar_kws={'label': 'Correlation'})
#         ax1.set_title('Asset Correlation Matrix')
        
#         # 2. VaR comparison
#         ax2 = axes[0, 1]
#         portfolios = []
#         var_values = []
#         cvar_values = []
        
#         if max_sharpe_portfolio['success']:
#             portfolios.append('Max Sharpe')
#             var_values.append(abs(sharpe_var))
#             cvar_values.append(abs(sharpe_cvar))
        
#         if min_vol_portfolio['success']:
#             portfolios.append('Min Volatility')
#             var_values.append(abs(minvol_var))
#             cvar_values.append(abs(minvol_cvar))
        
#         x_pos = np.arange(len(portfolios))
#         width = 0.35
        
#         ax2.bar(x_pos - width/2, var_values, width, label='VaR (5%)', alpha=0.7)
#         ax2.bar(x_pos + width/2, cvar_values, width, label='CVaR (5%)', alpha=0.7)
#         ax2.set_xlabel('Portfolio')
#         ax2.set_ylabel('Risk Measure')
#         ax2.set_title('Value at Risk Comparison')
#         ax2.set_xticks(x_pos)
#         ax2.set_xticklabels(portfolios)
#         ax2.legend()
#         ax2.grid(True, alpha=0.3)
        
#         # 3. Rolling volatility
#         ax3 = axes[1, 0]
#         rolling_vol = backtester.returns.rolling(window=30).std() * np.sqrt(252)
#         rolling_vol.mean(axis=1).plot(ax=ax3, label='Portfolio Volatility', alpha=0.7)
#         ax3.set_title('Rolling 30-Day Volatility (Annualized)')
#         ax3.set_ylabel('Volatility')
#         ax3.legend()
#         ax3.grid(True, alpha=0.3)
        
#         # 4. Stress test scenarios
#         ax4 = axes[1, 1]
        
#         # Define stress scenarios
#         stress_scenarios = {
#             'Market Crash': {asset: -0.20 for asset in backtester.returns.columns[:5]},
#             'Tech Selloff': {'AAPL': -0.30, 'MSFT': -0.25, 'GOOGL': -0.35, 'AMZN': -0.28},
#             'Financial Crisis': {'JPM': -0.40, 'BAC': -0.35, 'WFC': -0.30, 'GS': -0.45}
#         }
        
#         stress_results = risk_analyzer.stress_testing(stress_scenarios)
        
#         # Calculate portfolio impact for each scenario
#         scenario_impacts = []
#         scenario_names = []
        
#         for scenario_name, scenario_returns in stress_results.items():
#             if max_sharpe_portfolio['success']:
#                 impact = sum(max_sharpe_portfolio['weights'].get(asset, 0) * ret 
#                             for asset, ret in scenario_returns.items())
#                 scenario_impacts.append(abs(impact))
#                 scenario_names.append(scenario_name)
        
#         if scenario_impacts:
#             ax4.bar(scenario_names, scenario_impacts, alpha=0.7, color='red')
#             ax4.set_title('Stress Test Results (Max Sharpe Portfolio)')
#             ax4.set_ylabel('Portfolio Impact')
#             ax4.tick_params(axis='x', rotation=45)
#             ax4.grid(True, alpha=0.3)
        
#         plt.tight_layout()
#         plt.show()


# class ReportGenerator:
#     """Generate comprehensive backtesting reports"""
    
#     @staticmethod
#     def save_results(performance_df, max_sharpe_portfolio, min_vol_portfolio, 
#                     efficient_frontier, symbols, successful_symbols, period, price_data,
#                     sharpe_var=None, sharpe_cvar=None, minvol_var=None, minvol_cvar=None):
#         """Save backtesting results and generate comprehensive report"""
        
#         # Create output directory
#         output_dir = Path('./data/processed/backtesting')
#         output_dir.mkdir(parents=True, exist_ok=True)
        
#         # Save strategy performance
#         performance_df.to_csv(output_dir / 'strategy_performance.csv')
        
#         # Save portfolio weights
#         if max_sharpe_portfolio['success']:
#             pd.Series(max_sharpe_portfolio['weights']).to_csv(output_dir / 'max_sharpe_weights.csv')
        
#         if min_vol_portfolio['success']:
#             pd.Series(min_vol_portfolio['weights']).to_csv(output_dir / 'min_vol_weights.csv')
        
#         # Save efficient frontier
#         if not efficient_frontier.empty:
#             efficient_frontier.to_csv(output_dir / 'efficient_frontier.csv', index=False)
        
#         # Create comprehensive report
#         report = {
#             'analysis_date': datetime.now().isoformat(),
#             'universe': {
#                 'total_assets': len(successful_symbols),
#                 'sectors': list(symbols.keys()),
#                 'period': period,
#                 'date_range': {
#                     'start': price_data.index.min().isoformat(),
#                     'end': price_data.index.max().isoformat()
#                 }
#             },
#             'strategies_tested': list(performance_df.index),
#             'best_strategy': {
#                 'name': performance_df.sort_values('sharpe_ratio', ascending=False).index[0],
#                 'sharpe_ratio': float(performance_df['sharpe_ratio'].max()),
#                 'annual_return': float(performance_df.loc[performance_df['sharpe_ratio'].idxmax(), 'annual_return']),
#                 'max_drawdown': float(performance_df.loc[performance_df['sharpe_ratio'].idxmax(), 'max_drawdown'])
#             },
#             'portfolio_optimization': {
#                 'max_sharpe': {
#                     'success': max_sharpe_portfolio['success'],
#                     'expected_return': float(max_sharpe_portfolio.get('expected_return', 0)),
#                     'volatility': float(max_sharpe_portfolio.get('volatility', 0)),
#                     'sharpe_ratio': float(max_sharpe_portfolio.get('sharpe_ratio', 0))
#                 } if max_sharpe_portfolio['success'] else {'success': False},
#                 'min_volatility': {
#                     'success': min_vol_portfolio['success'],
#                     'expected_return': float(min_vol_portfolio.get('expected_return', 0)),
#                     'volatility': float(min_vol_portfolio.get('volatility', 0)),
#                     'sharpe_ratio': float(min_vol_portfolio.get('sharpe_ratio', 0))
#                 } if min_vol_portfolio['success'] else {'success': False}
#             },
#             'risk_analysis': {
#                 'max_sharpe_var': float(sharpe_var) if sharpe_var is not None else None,
#                 'max_sharpe_cvar': float(sharpe_cvar) if sharpe_cvar is not None else None,
#                 'min_vol_var': float(minvol_var) if minvol_var is not None else None,
#                 'min_vol_cvar': float(minvol_cvar) if minvol_cvar is not None else None
#             },
#             'key_insights': [
#                 f"Best performing strategy: {performance_df.sort_values('sharpe_ratio', ascending=False).index[0]}",
#                 f"Highest Sharpe ratio achieved: {performance_df['sharpe_ratio'].max():.3f}",
#                 f"Portfolio optimization {'successful' if max_sharpe_portfolio['success'] else 'failed'}",
#                 f"Risk analysis completed for {len(successful_symbols)} assets"
#             ]
#         }
        
#         # Save report
#         with open(output_dir / 'backtesting_report.json', 'w') as f:
#             json.dump(report, f, indent=2, default=str)
        
#         print(f"‚úÖ Backtesting results saved to {output_dir}")
        
#         return report
    
#     @staticmethod
#     def print_summary_report(report, max_sharpe_portfolio, sharpe_var=None):
#         """Print summary report to console"""
#         print("\nüìã Backtesting Summary Report")
#         print("=" * 50)
#         print(f"üìä Assets analyzed: {report['universe']['total_assets']}")
#         print(f"üìà Strategies tested: {len(report['strategies_tested'])}")
#         print(f"üèÜ Best strategy: {report['best_strategy']['name']}")
#         print(f"üìä Best Sharpe ratio: {report['best_strategy']['sharpe_ratio']:.3f}")
#         print(f"üìà Best annual return: {report['best_strategy']['annual_return']:.2%}")
        
#         if max_sharpe_portfolio['success']:
#             print(f"\nüéØ Optimal Portfolio (Max Sharpe):")
#             print(f"  Expected Return: {max_sharpe_portfolio['expected_return']:.2%}")
#             print(f"  Volatility: {max_sharpe_portfolio['volatility']:.2%}")
#             print(f"  Sharpe Ratio: {max_sharpe_portfolio['sharpe_ratio']:.3f}")
#             if sharpe_var is not None:
#                 print(f"  VaR (5%): {sharpe_var:.3f}")
        
#         print("\nüí° Key Recommendations:")
#         recommendations = [
#             "‚úÖ Diversification across sectors reduces portfolio risk",
#             "üìà Technical strategies show varying performance - combine multiple approaches",
#             "üéØ Portfolio optimization improves risk-adjusted returns",
#             "‚ö†Ô∏è  Regular rebalancing and risk monitoring essential",
#             "üìä Backtesting provides valuable insights but past performance doesn't guarantee future results"
#         ]
        
#         for rec in recommendations:
#             print(f"  {rec}")
        
#         print("\nüéâ Strategy backtesting pipeline completed successfully!")


# class InvestmentBacktestingPipeline:
#     """Main pipeline class that orchestrates the entire backtesting process"""
    
#     def __init__(self):
#         self.data_loader = None
#         self.price_data = None
#         self.strategies = {}
#         self.backtest_results = {}
#         self.performance_df = None
#         self.max_sharpe_portfolio = None
#         self.min_vol_portfolio = None
#         self.efficient_frontier = None
#         self.risk_analyzer = None
    
#     def run_full_pipeline(self, period='3y', initial_capital=100000, commission=0.001):
#         """Run the complete backtesting pipeline"""
        
#         print("üìä Investment Strategy Backtesting Pipeline Started")
#         print("=" * 50)
        
#         # 1. Load Market Data
#         print("\nüîÑ Step 1: Loading Market Data")
#         self.data_loader = MarketDataLoader()
#         self.price_data = self.data_loader.load_data(period)
        
#         if self.price_data.empty:
#             print("‚ùå No data available. Exiting.")
#             return
        
#         print(f"üìä Data loaded successfully: {self.price_data.shape}")
        
#         # 2. Generate Technical Strategies
#         print("\nüîÑ Step 2: Generating Technical Analysis Signals")
#         tech_strategies = TechnicalStrategies(self.price_data)
        
#         self.strategies = {
#             'MA_Crossover': tech_strategies.moving_average_crossover(20, 50),
#             'RSI_Strategy': tech_strategies.rsi_strategy(14, 30, 70),
#             'Bollinger_Bands': tech_strategies.bollinger_bands_strategy(20, 2),
#             'Momentum': tech_strategies.momentum_strategy(20, 5)
#         }
        
#         print(f"‚úÖ Generated signals for {len(self.strategies)} strategies")
#         for name, signals in self.strategies.items():
#             buy_signals = (signals == 1).sum().sum()
#             sell_signals = (signals == -1).sum().sum()
#             print(f"  {name}: {buy_signals} buy signals, {sell_signals} sell signals")
        
#         # 3. Visualize Strategy Signals
#         print("\nüîÑ Step 3: Visualizing Strategy Signals")
#         VisualizationManager.plot_strategy_signals(self.price_data, self.strategies)
        
#         # 4. Backtest Strategies
#         print("\nüîÑ Step 4: Running Strategy Backtests")
#         backtester = StrategyBacktester(self.price_data, initial_capital, commission)
        
#         benchmark_returns = self.price_data['SPY'].pct_change().dropna() if 'SPY' in self.price_data.columns else None
        
#         for strategy_name, signals in self.strategies.items():
#             print(f"Testing {strategy_name}...", end=' ')
            
#             try:
#                 result = backtester.backtest_strategy(signals, strategy_name)
#                 metrics = backtester.calculate_performance_metrics(
#                     result['strategy_returns'], benchmark_returns
#                 )
                
#                 self.backtest_results[strategy_name] = {
#                     'result': result,
#                     'metrics': metrics
#                 }
                
#                 print(f"‚úÖ Return: {metrics['annual_return']:.2%}, Sharpe: {metrics['sharpe_ratio']:.2f}")
            
#             except Exception as e:
#                 print(f"‚ùå Error: {e}")
        
#         # Create performance summary
#         self.performance_df = pd.DataFrame({
#             name: results['metrics'] 
#             for name, results in self.backtest_results.items()
#         }).T
        
#         print("\nüìä Strategy Performance Summary:")
#         print(self.performance_df.round(4))
        
#         # 5. Visualize Strategy Performance
#         print("\nüîÑ Step 5: Visualizing Strategy Performance")
#         VisualizationManager.plot_strategy_performance(self.backtest_results, benchmark_returns)
        
#         # 6. Portfolio Optimization
#         print("\nüîÑ Step 6: Portfolio Optimization")
#         optimizer = PortfolioOptimizer(backtester.returns)
        
#         # Find optimal portfolios
#         print("Finding optimal portfolios...")
        
#         # Maximum Sharpe ratio portfolio
#         self.max_sharpe_portfolio = optimizer.optimize_portfolio('sharpe')
#         print(f"‚úÖ Max Sharpe portfolio: {self.max_sharpe_portfolio.get('sharpe_ratio', 0):.3f}")
        
#         # Minimum volatility portfolio
#         self.min_vol_portfolio = optimizer.optimize_portfolio('volatility')
#         print(f"‚úÖ Min volatility portfolio: {self.min_vol_portfolio.get('volatility', 0):.3f}")
        
#         # Generate efficient frontier
#         print("Generating efficient frontier...")
#         self.efficient_frontier = optimizer.efficient_frontier(30)
        
#         # Display optimal portfolios
#         if self.max_sharpe_portfolio['success']:
#             print("\nüìä Maximum Sharpe Ratio Portfolio:")
#             sharpe_weights = pd.Series(self.max_sharpe_portfolio['weights']).sort_values(ascending=False)
#             print(f"Expected Return: {self.max_sharpe_portfolio['expected_return']:.2%}")
#             print(f"Volatility: {self.max_sharpe_portfolio['volatility']:.2%}")
#             print(f"Sharpe Ratio: {self.max_sharpe_portfolio['sharpe_ratio']:.3f}")
#             print("\nTop 10 Holdings:")
#             print(sharpe_weights.head(10).apply(lambda x: f"{x:.1%}"))
        
#         if self.min_vol_portfolio['success']:
#             print("\nüìä Minimum Volatility Portfolio:")
#             min_vol_weights = pd.Series(self.min_vol_portfolio['weights']).sort_values(ascending=False)
#             print(f"Expected Return: {self.min_vol_portfolio['expected_return']:.2%}")
#             print(f"Volatility: {self.min_vol_portfolio['volatility']:.2%}")
#             print(f"Sharpe Ratio: {self.min_vol_portfolio['sharpe_ratio']:.3f}")
#             print("\nTop 10 Holdings:")
#             print(min_vol_weights.head(10).apply(lambda x: f"{x:.1%}"))
        
#         # 7. Visualize Portfolio Optimization
#         print("\nüîÑ Step 7: Visualizing Portfolio Optimization")
#         VisualizationManager.plot_portfolio_optimization(
#             optimizer, self.max_sharpe_portfolio, self.min_vol_portfolio, self.efficient_frontier
#         )
        
#         # 8. Risk Analysis
#         print("\nüîÑ Step 8: Performing Risk Analysis")
#         self.risk_analyzer = RiskAnalyzer(backtester.returns)
        
#         # Calculate risk metrics for optimal portfolios
#         sharpe_var = sharpe_cvar = minvol_var = minvol_cvar = None
        
#         if self.max_sharpe_portfolio['success']:
#             sharpe_weights_array = np.array([self.max_sharpe_portfolio['weights'].get(col, 0) 
#                                             for col in backtester.returns.columns])
            
#             sharpe_var = self.risk_analyzer.calculate_var(0.05, sharpe_weights_array)
#             sharpe_cvar = self.risk_analyzer.calculate_cvar(0.05, sharpe_weights_array)
            
#             print(f"üìä Max Sharpe Portfolio Risk Metrics:")
#             print(f"  VaR (5%): {sharpe_var:.3f}")
#             print(f"  CVaR (5%): {sharpe_cvar:.3f}")
        
#         if self.min_vol_portfolio['success']:
#             minvol_weights_array = np.array([self.min_vol_portfolio['weights'].get(col, 0) 
#                                             for col in backtester.returns.columns])
            
#             minvol_var = self.risk_analyzer.calculate_var(0.05, minvol_weights_array)
#             minvol_cvar = self.risk_analyzer.calculate_cvar(0.05, minvol_weights_array)
            
#             print(f"\nüìä Min Volatility Portfolio Risk Metrics:")
#             print(f"  VaR (5%): {minvol_var:.3f}")
#             print(f"  CVaR (5%): {minvol_cvar:.3f}")
        
#         # Correlation analysis
#         corr_analysis = self.risk_analyzer.correlation_analysis()
        
#         print("\nüîó Correlation Analysis:")
#         print("\nHighest Correlations:")
#         print(corr_analysis['highest_correlations'].head())
        
#         print("\nLowest Correlations:")
#         print(corr_analysis['lowest_correlations'].head())
        
#         # 9. Visualize Risk Analysis
#         print("\nüîÑ Step 9: Visualizing Risk Analysis")
#         VisualizationManager.plot_risk_analysis(
#             self.risk_analyzer, self.max_sharpe_portfolio, self.min_vol_portfolio,
#             sharpe_var, sharpe_cvar, minvol_var, minvol_cvar, backtester
#         )
        
#         # 10. Generate and Save Report
#         print("\nüîÑ Step 10: Generating Report")
#         report = ReportGenerator.save_results(
#             self.performance_df, self.max_sharpe_portfolio, self.min_vol_portfolio,
#             self.efficient_frontier, self.data_loader.symbols, self.data_loader.successful_symbols,
#             period, self.price_data, sharpe_var, sharpe_cvar, minvol_var, minvol_cvar
#         )
        
#         ReportGenerator.print_summary_report(report, self.max_sharpe_portfolio, sharpe_var)
        
#         return {
#             'data_loader': self.data_loader,
#             'price_data': self.price_data,
#             'strategies': self.strategies,
#             'backtest_results': self.backtest_results,
#             'performance_df': self.performance_df,
#             'max_sharpe_portfolio': self.max_sharpe_portfolio,
#             'min_vol_portfolio': self.min_vol_portfolio,
#             'efficient_frontier': self.efficient_frontier,
#             'risk_analyzer': self.risk_analyzer,
#             'report': report
#         }


# def main():
#     """Main function to run the investment backtesting pipeline"""
    
#     try:
#         # Initialize and run the pipeline
#         pipeline = InvestmentBacktestingPipeline()
#         results = pipeline.run_full_pipeline(
#             period='3y',           # Analysis period
#             initial_capital=100000, # Starting capital
#             commission=0.001       # Transaction cost (0.1%)
#         )
        
#         print("\n‚úÖ Pipeline completed successfully!")
        
#         # Optional: Return results for further analysis
#         return results
        
#     except KeyboardInterrupt:
#         print("\n‚ùå Pipeline interrupted by user")
#     except Exception as e:
#         print(f"\n‚ùå Pipeline failed with error: {e}")
#         import traceback
#         traceback.print_exc()


# if __name__ == "__main__":
#     # Run the main pipeline
#     results = main()
    
#     # Optional: Access specific results
#     if results:
#         print(f"\nüìä Final Results Available:")
#         print(f"  - Price data shape: {results['price_data'].shape}")
#         print(f"  - Strategies tested: {len(results['strategies'])}")
#         print(f"  - Best strategy: {results['performance_df'].sort_values('sharpe_ratio', ascending=False).index[0]}")
#         print(f"  - Portfolio optimization: {'‚úÖ Success' if results['max_sharpe_portfolio']['success'] else '‚ùå Failed'}")
        
#         # Example of accessing specific results:
#         # best_strategy = results['performance_df'].sort_values('sharpe_ratio', ascending=False).iloc[0]
#         # optimal_portfolio = results['max_sharpe_portfolio']
#         # correlation_matrix = results['risk_analyzer'].correlation_analysis()['correlation_matrix']


#!/usr/bin/env python3
"""
Investment Strategy Backtesting for FinDocGPT

This script demonstrates comprehensive backtesting of investment strategies including:
- Technical analysis strategies
- Sentiment-based strategies  
- Portfolio optimization
- Risk management and performance metrics

Author: FinDocGPT Team
Date: 2025
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
import warnings
warnings.filterwarnings('ignore')

# Financial analysis
import yfinance as yf
from scipy import stats
from scipy.optimize import minimize

# Portfolio optimization
try:
    from pypfopt import EfficientFrontier, risk_models, expected_returns
    pypfopt_available = True
except ImportError:
    pypfopt_available = False
    print("PyPortfolioOpt not available - using custom implementation")

# Additional imports for report generation
from pathlib import Path
import json


class MarketDataLoader:
    """Class to handle market data loading and preparation"""
    
    def __init__(self):
        self.symbols = {
            'Tech': ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'TSLA', 'NVDA'],
            'Finance': ['JPM', 'BAC', 'WFC', 'GS', 'C'],
            'Healthcare': ['JNJ', 'PFE', 'UNH', 'ABBV'],
            'Energy': ['XOM', 'CVX', 'COP'],
            'ETFs': ['SPY', 'QQQ', 'IWM']
        }
        self.all_symbols = [symbol for sector_symbols in self.symbols.values() 
                           for symbol in sector_symbols]
        self.market_data = {}
        self.price_data = pd.DataFrame()
        self.successful_symbols = []
        self.failed_symbols = []
    
    def load_data(self, period='3y'):
        """Load market data for all symbols"""
        print(f"üåç Investment Universe: {len(self.all_symbols)} symbols across {len(self.symbols)} sectors")
        print(f"üìÖ Analysis period: {period}")
        
        print("\nüìä Downloading market data...")
        
        for symbol in self.all_symbols:
            try:
                data = yf.download(symbol, period=period, progress=False)
                if not data.empty and len(data) > 100:  # Minimum data requirement
                    self.market_data[symbol] = data
                    self.successful_symbols.append(symbol)
                    print(f"‚úÖ {symbol}: {len(data)} days")
                else:
                    self.failed_symbols.append(symbol)
                    print(f"‚ùå {symbol}: Insufficient data")
            except Exception as e:
                self.failed_symbols.append(symbol)
                print(f"‚ùå {symbol}: Error - {e}")
        
        print(f"\n‚úÖ Successfully loaded data for {len(self.successful_symbols)} symbols")
        if self.failed_symbols:
            print(f"‚ùå Failed to load: {self.failed_symbols}")
        
        # Create combined price DataFrame
        if self.market_data:
            # Get the first symbol to establish the index
            first_symbol = list(self.market_data.keys())[0]
            first_data = self.market_data[first_symbol]
            
            # Create DataFrame with proper index
            self.price_data = pd.DataFrame(index=first_data.index)
            
            # Add each symbol's close price
            for symbol, data in self.market_data.items():
                self.price_data[symbol] = data['Close']
            
            # Drop rows with any NaN values
            self.price_data = self.price_data.dropna()
        
        print(f"\nüìä Combined dataset: {self.price_data.shape}")
        print(f"üìÖ Date range: {self.price_data.index.min().date()} to {self.price_data.index.max().date()}")
        
        return self.price_data


class TechnicalStrategies:
    """Implementation of various technical analysis strategies"""
    
    def __init__(self, price_data):
        self.price_data = price_data
        self.returns = price_data.pct_change().dropna()
    
    def moving_average_crossover(self, short_window=20, long_window=50):
        """Moving average crossover strategy"""
        signals = pd.DataFrame(index=self.price_data.index, columns=self.price_data.columns)
        
        for symbol in self.price_data.columns:
            prices = self.price_data[symbol].dropna()
            
            # Calculate moving averages
            short_ma = prices.rolling(window=short_window).mean()
            long_ma = prices.rolling(window=long_window).mean()
            
            # Generate signals
            signal = pd.Series(0, index=prices.index)
            signal[short_ma > long_ma] = 1  # Buy signal
            signal[short_ma < long_ma] = -1  # Sell signal
            
            # Align with main index
            signals[symbol] = signal.reindex(self.price_data.index).fillna(0)
        
        return signals.fillna(0)
    
    def rsi_strategy(self, window=14, oversold=30, overbought=70):
        """RSI-based mean reversion strategy"""
        signals = pd.DataFrame(index=self.price_data.index, columns=self.price_data.columns)
        
        for symbol in self.price_data.columns:
            prices = self.price_data[symbol].dropna()
            
            # Calculate RSI
            delta = prices.diff()
            gain = (delta.where(delta > 0, 0)).rolling(window=window).mean()
            loss = (-delta.where(delta < 0, 0)).rolling(window=window).mean()
            
            # Avoid division by zero
            rs = gain / loss.replace(0, np.nan)
            rsi = 100 - (100 / (1 + rs))
            
            # Generate signals
            signal = pd.Series(0, index=prices.index)
            signal[rsi < oversold] = 1  # Buy when oversold
            signal[rsi > overbought] = -1  # Sell when overbought
            
            # Align with main index
            signals[symbol] = signal.reindex(self.price_data.index).fillna(0)
        
        return signals.fillna(0)
    
    def bollinger_bands_strategy(self, window=20, num_std=2):
        """Bollinger Bands mean reversion strategy"""
        signals = pd.DataFrame(index=self.price_data.index, columns=self.price_data.columns)
        
        for symbol in self.price_data.columns:
            prices = self.price_data[symbol].dropna()
            
            # Calculate Bollinger Bands
            sma = prices.rolling(window=window).mean()
            std = prices.rolling(window=window).std()
            upper_band = sma + (std * num_std)
            lower_band = sma - (std * num_std)
            
            # Generate signals
            signal = pd.Series(0, index=prices.index)
            signal[prices < lower_band] = 1  # Buy when below lower band
            signal[prices > upper_band] = -1  # Sell when above upper band
            
            # Align with main index
            signals[symbol] = signal.reindex(self.price_data.index).fillna(0)
        
        return signals.fillna(0)
    
    def momentum_strategy(self, lookback=20, hold_period=5):
        """Momentum strategy based on recent returns"""
        signals = pd.DataFrame(index=self.price_data.index, columns=self.price_data.columns)
        
        for symbol in self.price_data.columns:
            prices = self.price_data[symbol].dropna()
            
            # Calculate momentum (return over lookback period)
            momentum = prices.pct_change(periods=lookback)
            
            # Generate signals based on momentum percentiles
            signal = pd.Series(0, index=prices.index)
            
            # Buy top 30% momentum, sell bottom 30%
            for i, date in enumerate(momentum.index):
                if pd.notna(momentum[date]) and i >= 60:  # Need sufficient history
                    # Use rolling window for dynamic thresholds
                    window_data = momentum.iloc[max(0, i-60):i]
                    if len(window_data) > 10:  # Ensure sufficient data
                        rolling_quantile_high = window_data.quantile(0.7)
                        rolling_quantile_low = window_data.quantile(0.3)
                        
                        if pd.notna(rolling_quantile_high) and momentum[date] > rolling_quantile_high:
                            signal[date] = 1
                        elif pd.notna(rolling_quantile_low) and momentum[date] < rolling_quantile_low:
                            signal[date] = -1
            
            # Align with main index
            signals[symbol] = signal.reindex(self.price_data.index).fillna(0)
        
        return signals.fillna(0)


class StrategyBacktester:
    """Comprehensive backtesting engine"""
    
    def __init__(self, price_data, initial_capital=100000, commission=0.001):
        self.price_data = price_data
        self.returns = price_data.pct_change().dropna()
        self.initial_capital = initial_capital
        self.commission = commission
    
    def backtest_strategy(self, signals, strategy_name):
        """Backtest a single strategy"""
        
        # Align signals with returns data (both index and columns)
        aligned_signals = signals.reindex(index=self.returns.index, columns=self.returns.columns).fillna(0)
        
        # Initialize portfolio tracking
        portfolio_value = pd.Series(self.initial_capital, index=self.returns.index)
        positions = pd.DataFrame(0, index=self.returns.index, columns=self.returns.columns)
        trades = []
        cash = self.initial_capital
        
        for i, date in enumerate(self.returns.index):
            if i == 0:
                continue
            
            prev_date = self.returns.index[i-1]
            current_positions = positions.loc[prev_date].copy()
            
            # Update portfolio value based on returns
            if i > 1:
                position_values = current_positions * self.price_data.loc[date]
                position_values = position_values.fillna(0)
                portfolio_value[date] = cash + position_values.sum()
            
            # Process signals for each asset
            for symbol in self.returns.columns:
                if symbol not in aligned_signals.columns:
                    continue
                    
                signal = aligned_signals.loc[date, symbol]
                current_pos = current_positions[symbol]
                
                if symbol not in self.price_data.columns or pd.isna(self.price_data.loc[date, symbol]):
                    continue
                    
                price = self.price_data.loc[date, symbol]
                
                if pd.isna(signal) or signal == 0:
                    continue
                
                # Calculate position size (equal weight among active positions)
                active_signals = aligned_signals.loc[date]
                num_active = (active_signals == 1).sum()
                
                if signal == 1 and current_pos == 0 and num_active > 0:  # Buy signal
                    position_value = max(portfolio_value[date], cash) / num_active * 0.95  # 95% of available capital
                    shares = position_value / price
                    cost = shares * price * (1 + self.commission)
                    
                    if cost <= cash and cost > 0:
                        current_positions[symbol] = shares
                        cash -= cost
                        trades.append({
                            'date': date,
                            'symbol': symbol,
                            'action': 'BUY',
                            'shares': shares,
                            'price': price,
                            'value': cost
                        })
                
                elif signal == -1 and current_pos > 0:  # Sell signal
                    proceeds = current_pos * price * (1 - self.commission)
                    cash += proceeds
                    trades.append({
                        'date': date,
                        'symbol': symbol,
                        'action': 'SELL',
                        'shares': current_pos,
                        'price': price,
                        'value': proceeds
                    })
                    current_positions[symbol] = 0
            
            positions.loc[date] = current_positions
        
        # Calculate final portfolio value
        if len(positions) > 0:
            final_positions = positions.iloc[-1]
            final_prices = self.price_data.iloc[-1]
            # Handle NaN values in final calculation
            final_position_value = (final_positions * final_prices).fillna(0).sum()
            final_portfolio_value = cash + final_position_value
            portfolio_value.iloc[-1] = final_portfolio_value
        else:
            final_portfolio_value = self.initial_capital
        
        # Calculate strategy returns
        strategy_returns = portfolio_value.pct_change().dropna()
        
        return {
            'portfolio_value': portfolio_value,
            'strategy_returns': strategy_returns,
            'positions': positions,
            'trades': trades,
            'final_value': final_portfolio_value
        }
    
    def calculate_performance_metrics(self, strategy_returns, benchmark_returns=None):
        """Calculate comprehensive performance metrics"""
        
        if len(strategy_returns) == 0 or strategy_returns.isna().all():
            # Return default metrics if no valid returns
            return {
                'total_return': 0,
                'annual_return': 0,
                'volatility': 0,
                'sharpe_ratio': 0,
                'max_drawdown': 0,
                'win_rate': 0,
                'total_trades': 0,
                'calmar_ratio': 0
            }
        
        # Clean returns data
        clean_returns = strategy_returns.dropna()
        if len(clean_returns) == 0:
            return {
                'total_return': 0,
                'annual_return': 0,
                'volatility': 0,
                'sharpe_ratio': 0,
                'max_drawdown': 0,
                'win_rate': 0,
                'total_trades': 0,
                'calmar_ratio': 0
            }
        
        # Basic metrics
        total_return = (clean_returns + 1).prod() - 1
        annual_return = (1 + total_return) ** (252 / len(clean_returns)) - 1 if len(clean_returns) > 0 else 0
        volatility = clean_returns.std() * np.sqrt(252)
        
        # Risk metrics
        sharpe_ratio = annual_return / volatility if volatility > 0 else 0
        
        # Drawdown analysis
        cumulative_returns = (1 + clean_returns).cumprod()
        peak = cumulative_returns.expanding().max()
        drawdown = (cumulative_returns - peak) / peak
        max_drawdown = abs(drawdown.min()) if len(drawdown) > 0 else 0
        
        # Additional metrics
        positive_days = (clean_returns > 0).sum()
        total_days = len(clean_returns)
        win_rate = positive_days / total_days if total_days > 0 else 0
        
        metrics = {
            'total_return': float(total_return),
            'annual_return': float(annual_return),
            'volatility': float(volatility),
            'sharpe_ratio': float(sharpe_ratio),
            'max_drawdown': float(max_drawdown),
            'win_rate': float(win_rate),
            'total_trades': total_days,
            'calmar_ratio': float(annual_return / max_drawdown if max_drawdown > 0 else 0)
        }
        
        # Benchmark comparison
        if benchmark_returns is not None and len(benchmark_returns) > 0:
            clean_benchmark = benchmark_returns.dropna()
            if len(clean_benchmark) > 0:
                benchmark_total = (clean_benchmark + 1).prod() - 1
                benchmark_annual = (1 + benchmark_total) ** (252 / len(clean_benchmark)) - 1
                excess_return = annual_return - benchmark_annual
                
                # Beta calculation
                aligned_strategy = clean_returns.reindex(clean_benchmark.index).dropna()
                aligned_benchmark = clean_benchmark.reindex(aligned_strategy.index).dropna()
                
                if len(aligned_strategy) > 0 and len(aligned_benchmark) > 0:
                    covariance = np.cov(aligned_strategy, aligned_benchmark)[0][1]
                    benchmark_variance = np.var(aligned_benchmark)
                    beta = covariance / benchmark_variance if benchmark_variance > 0 else 0
                    
                    metrics.update({
                        'benchmark_annual_return': float(benchmark_annual),
                        'excess_return': float(excess_return),
                        'beta': float(beta),
                        'alpha': float(annual_return - (0.02 + beta * (benchmark_annual - 0.02)))  # Assuming 2% risk-free rate
                    })
        
        return metrics


class PortfolioOptimizer:
    """Portfolio optimization using Modern Portfolio Theory"""
    
    def __init__(self, returns_data):
        self.returns = returns_data.dropna()
        # Remove columns with all NaN or zero variance
        self.returns = self.returns.loc[:, self.returns.std() > 1e-6]
        
        if len(self.returns.columns) == 0:
            raise ValueError("No valid assets with sufficient variance for optimization")
            
        self.mean_returns = self.returns.mean() * 252
        self.cov_matrix = self.returns.cov() * 252
        self.n_assets = len(self.returns.columns)
    
    def portfolio_performance(self, weights):
        """Calculate portfolio performance metrics"""
        portfolio_return = np.dot(weights, self.mean_returns)
        portfolio_volatility = np.sqrt(np.dot(weights.T, np.dot(self.cov_matrix, weights)))
        sharpe_ratio = portfolio_return / portfolio_volatility if portfolio_volatility > 0 else 0
        return portfolio_return, portfolio_volatility, sharpe_ratio
    
    def optimize_portfolio(self, objective='sharpe', target_return=None):
        """Optimize portfolio based on objective"""
        
        if self.n_assets < 2:
            return {'success': False, 'message': 'Need at least 2 assets for optimization'}
        
        # Constraints
        constraints = [{'type': 'eq', 'fun': lambda x: np.sum(x) - 1}]  # Weights sum to 1
        
        if target_return is not None:
            constraints.append({
                'type': 'eq', 
                'fun': lambda x: np.dot(x, self.mean_returns) - target_return
            })
        
        # Bounds (no short selling)
        bounds = tuple((0, 1) for _ in range(self.n_assets))
        
        # Initial guess
        x0 = np.array([1/self.n_assets] * self.n_assets)
        
        # Objective functions
        if objective == 'sharpe':
            def objective_func(weights):
                _, vol, sharpe = self.portfolio_performance(weights)
                return -sharpe  # Negative because we're minimizing
        elif objective == 'volatility':
            def objective_func(weights):
                _, vol, _ = self.portfolio_performance(weights)
                return vol
        else:
            raise ValueError("Objective must be 'sharpe' or 'volatility'")
        
        # Optimize with multiple attempts
        best_result = None
        
        for attempt in range(3):  # Try multiple starting points
            try:
                # Vary initial guess slightly
                x0_attempt = x0 + np.random.normal(0, 0.01, self.n_assets)
                x0_attempt = np.abs(x0_attempt)
                x0_attempt = x0_attempt / x0_attempt.sum()  # Normalize
                
                result = minimize(objective_func, x0_attempt, method='SLSQP', 
                                bounds=bounds, constraints=constraints,
                                options={'maxiter': 1000})
                
                if result.success and (best_result is None or result.fun < best_result.fun):
                    best_result = result
            except:
                continue
        
        if best_result and best_result.success:
            optimal_weights = best_result.x
            ret, vol, sharpe = self.portfolio_performance(optimal_weights)
            
            return {
                'weights': dict(zip(self.returns.columns, optimal_weights)),
                'expected_return': float(ret),
                'volatility': float(vol),
                'sharpe_ratio': float(sharpe),
                'success': True
            }
        else:
            return {'success': False, 'message': 'Optimization failed'}
    
    def efficient_frontier(self, num_points=50):
        """Generate efficient frontier"""
        if self.n_assets < 2:
            return pd.DataFrame()
            
        min_ret = self.mean_returns.min()
        max_ret = self.mean_returns.max()
        target_returns = np.linspace(min_ret, max_ret, num_points)
        
        efficient_portfolios = []
        
        for target_ret in target_returns:
            try:
                result = self.optimize_portfolio('volatility', target_ret)
                if result.get('success'):
                    efficient_portfolios.append({
                        'return': result['expected_return'],
                        'volatility': result['volatility'],
                        'sharpe': result['sharpe_ratio']
                    })
            except:
                continue
        
        return pd.DataFrame(efficient_portfolios)


class RiskAnalyzer:
    """Comprehensive risk analysis"""
    
    def __init__(self, returns_data):
        self.returns = returns_data.dropna()
        # Remove columns with insufficient data
        self.returns = self.returns.loc[:, self.returns.count() > len(self.returns) * 0.5]
    
    def calculate_var(self, confidence_level=0.05, portfolio_weights=None):
        """Calculate Value at Risk"""
        if portfolio_weights is not None:
            # Ensure weights align with available columns
            aligned_weights = pd.Series(portfolio_weights, index=self.returns.columns).fillna(0)
            portfolio_returns = (self.returns * aligned_weights).sum(axis=1)
        else:
            # Equal weight portfolio
            portfolio_returns = self.returns.mean(axis=1)
        
        portfolio_returns = portfolio_returns.dropna()
        if len(portfolio_returns) == 0:
            return 0
            
        var = np.percentile(portfolio_returns, confidence_level * 100)
        return float(var)
    
    def calculate_cvar(self, confidence_level=0.05, portfolio_weights=None):
        """Calculate Conditional Value at Risk (Expected Shortfall)"""
        if portfolio_weights is not None:
            aligned_weights = pd.Series(portfolio_weights, index=self.returns.columns).fillna(0)
            portfolio_returns = (self.returns * aligned_weights).sum(axis=1)
        else:
            portfolio_returns = self.returns.mean(axis=1)
        
        portfolio_returns = portfolio_returns.dropna()
        if len(portfolio_returns) == 0:
            return 0
        
        var = self.calculate_var(confidence_level, portfolio_weights)
        cvar_returns = portfolio_returns[portfolio_returns <= var]
        
        if len(cvar_returns) == 0:
            return float(var)
        
        cvar = cvar_returns.mean()
        return float(cvar)
    
    def correlation_analysis(self):
        """Analyze asset correlations"""
        if len(self.returns.columns) < 2:
            return {
                'correlation_matrix': pd.DataFrame(),
                'highest_correlations': pd.DataFrame(),
                'lowest_correlations': pd.DataFrame()
            }
            
        correlation_matrix = self.returns.corr()
        
        # Find highest and lowest correlations
        corr_pairs = []
        for i in range(len(correlation_matrix.columns)):
            for j in range(i+1, len(correlation_matrix.columns)):
                corr_val = correlation_matrix.iloc[i, j]
                if not pd.isna(corr_val):
                    corr_pairs.append({
                        'asset1': correlation_matrix.columns[i],
                        'asset2': correlation_matrix.columns[j],
                        'correlation': corr_val
                    })
        
        if corr_pairs:
            corr_df = pd.DataFrame(corr_pairs).sort_values('correlation', ascending=False)
            highest_corr = corr_df.head(10)
            lowest_corr = corr_df.tail(10)
        else:
            highest_corr = pd.DataFrame()
            lowest_corr = pd.DataFrame()
        
        return {
            'correlation_matrix': correlation_matrix,
            'highest_correlations': highest_corr,
            'lowest_correlations': lowest_corr
        }
    
    def stress_testing(self, scenarios):
        """Perform stress testing on portfolios"""
        stress_results = {}
        
        for scenario_name, scenario_returns in scenarios.items():
            # Apply scenario to each asset
            stressed_returns = {}
            for asset in self.returns.columns:
                if asset in scenario_returns:
                    stressed_returns[asset] = scenario_returns[asset]
                else:
                    # Default stress (e.g., -10%)
                    stressed_returns[asset] = -0.10
            
            stress_results[scenario_name] = stressed_returns
        
        return stress_results


class VisualizationManager:
    """Handle all visualization tasks"""
    
    @staticmethod
    def plot_strategy_signals(price_data, strategies, sample_symbol=None):
        """Visualize strategy signals for a sample stock"""
        if price_data.empty or len(strategies) == 0:
            print("No data available for signal visualization")
            return
            
        if sample_symbol is None:
            sample_symbol = 'AAPL' if 'AAPL' in price_data.columns else price_data.columns[0]
        
        sample_data = price_data[sample_symbol].tail(252)  # Last year
        
        # Determine number of subplots needed
        n_strategies = len(strategies)
        n_rows = (n_strategies + 1) // 2
        n_cols = 2 if n_strategies > 1 else 1
        
        fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5*n_rows))
        if n_strategies == 1:
            axes = [axes]
        elif n_rows == 1:
            axes = axes
        else:
            axes = axes.flatten()
        
        for i, (strategy_name, signals) in enumerate(strategies.items()):
            if i >= len(axes):
                break
                
            ax = axes[i]
            
            # Plot price
            ax.plot(sample_data.index, sample_data.values, label='Price', alpha=0.7)
            
            # Plot buy/sell signals
            if sample_symbol in signals.columns:
                strategy_signals = signals[sample_symbol].tail(252)
                buy_dates = strategy_signals[strategy_signals == 1].index
                sell_dates = strategy_signals[strategy_signals == -1].index
                
                if len(buy_dates) > 0:
                    ax.scatter(buy_dates, sample_data[buy_dates], 
                              color='green', marker='^', s=50, label='Buy Signal')
                
                if len(sell_dates) > 0:
                    ax.scatter(sell_dates, sample_data[sell_dates], 
                              color='red', marker='v', s=50, label='Sell Signal')
            
            ax.set_title(f'{strategy_name} - {sample_symbol}')
            ax.legend()
            ax.grid(True, alpha=0.3)
        
        # Hide empty subplots
        for j in range(i+1, len(axes)):
            axes[j].set_visible(False)
        
        plt.tight_layout()
        plt.show()
    
    @staticmethod
    def plot_strategy_performance(backtest_results, benchmark_returns=None):
        """Visualize strategy performance"""
        if not backtest_results:
            print("No backtest results available for visualization")
            return
            
        fig, axes = plt.subplots(2, 2, figsize=(15, 10))
        
        # Create performance dataframe
        performance_df = pd.DataFrame({
            name: results['metrics'] 
            for name, results in backtest_results.items()
        }).T
        
        if performance_df.empty:
            print("No valid performance metrics to visualize")
            return
        
        # 1. Cumulative returns
        ax1 = axes[0, 0]
        for strategy_name, results in backtest_results.items():
            portfolio_value = results['result']['portfolio_value']
            if not portfolio_value.empty:
                cumulative_return = portfolio_value / portfolio_value.iloc[0]
                ax1.plot(cumulative_return.index, cumulative_return.values, 
                        label=strategy_name, linewidth=2)
        
        # Add benchmark if available
        if benchmark_returns is not None and len(benchmark_returns) > 0:
            benchmark_cumulative = (1 + benchmark_returns).cumprod()
            ax1.plot(benchmark_cumulative.index, benchmark_cumulative.values, 
                    label='SPY Benchmark', color='black', linestyle='--', alpha=0.7)
        
        ax1.set_title('Cumulative Returns Comparison')
        ax1.legend()
        ax1.grid(True, alpha=0.3)
        
        # 2. Annual returns vs volatility
        ax2 = axes[0, 1]
        if 'annual_return' in performance_df.columns and 'volatility' in performance_df.columns:
            returns = performance_df['annual_return'].values
            volatilities = performance_df['volatility'].values
            sharpe_ratios = performance_df.get('sharpe_ratio', pd.Series(0, index=performance_df.index)).values
            
            scatter = ax2.scatter(volatilities, returns, s=100, c=sharpe_ratios, 
                                 cmap='RdYlGn', alpha=0.7)
            ax2.set_xlabel('Volatility (Annual)')
            ax2.set_ylabel('Annual Return')
            ax2.set_title('Risk-Return Profile')
            ax2.grid(True, alpha=0.3)
            
            # Add strategy labels
            for i, strategy in enumerate(performance_df.index):
                ax2.annotate(strategy, (volatilities[i], returns[i]), 
                            xytext=(5, 5), textcoords='offset points', fontsize=9)
            
            plt.colorbar(scatter, ax=ax2, label='Sharpe Ratio')
        
        # 3. Drawdown analysis
        ax3 = axes[1, 0]
        for strategy_name, results in backtest_results.items():
            portfolio_value = results['result']['portfolio_value']
            if not portfolio_value.empty:
                cumulative = portfolio_value / portfolio_value.iloc[0]
                peak = cumulative.expanding().max()
                drawdown = (cumulative - peak) / peak
                ax3.fill_between(drawdown.index, drawdown.values, 0, alpha=0.3, label=strategy_name)
        
        ax3.set_title('Strategy Drawdowns')
        ax3.set_ylabel('Drawdown')
        ax3.legend()
        ax3.grid(True, alpha=0.3)
        
        # 4. Performance metrics comparison
        ax4 = axes[1, 1]
        if len(performance_df) > 0:
            metrics_to_plot = []
            for metric in ['annual_return', 'sharpe_ratio', 'max_drawdown']:
                if metric in performance_df.columns:
                    metrics_to_plot.append(metric)
            
            if metrics_to_plot:
                x_pos = np.arange(len(performance_df.index))
                width = 0.25
                
                for i, metric in enumerate(metrics_to_plot):
                    values = performance_df[metric].values
                    if metric == 'max_drawdown':
                        values = -values  # Make negative for better visualization
                    ax4.bar(x_pos + i*width, values, width, label=metric.replace('_', ' ').title())
                
                ax4.set_xlabel('Strategy')
                ax4.set_ylabel('Value')
                ax4.set_title('Key Performance Metrics')
                ax4.set_xticks(x_pos + width)
                ax4.set_xticklabels(performance_df.index, rotation=45)
                ax4.legend()
                ax4.grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.show()
    
    @staticmethod
    def plot_portfolio_optimization(optimizer, max_sharpe_portfolio, min_vol_portfolio, efficient_frontier):
        """Visualize portfolio optimization results"""
        fig, axes = plt.subplots(1, 2, figsize=(15, 6))
        
        # 1. Efficient frontier
        ax1 = axes[0]
        if not efficient_frontier.empty:
            ax1.plot(efficient_frontier['volatility'], efficient_frontier['return'], 
                    'b-', linewidth=2, label='Efficient Frontier')
        
        # Plot optimal portfolios
        if max_sharpe_portfolio.get('success', False):
            ax1.scatter(max_sharpe_portfolio['volatility'], max_sharpe_portfolio['expected_return'],
                       color='red', s=100, label='Max Sharpe', marker='*', zorder=5)
        
        if min_vol_portfolio.get('success', False):
            ax1.scatter(min_vol_portfolio['volatility'], min_vol_portfolio['expected_return'],
                       color='green', s=100, label='Min Volatility', marker='o', zorder=5)
        
        # Plot individual assets
        try:
            for asset in optimizer.returns.columns[:10]:  # Show first 10 assets
                ret = optimizer.mean_returns[asset]
                vol = np.sqrt(optimizer.cov_matrix.loc[asset, asset])
                ax1.scatter(vol, ret, alpha=0.5, s=30)
                ax1.annotate(asset, (vol, ret), xytext=(5, 5), 
                            textcoords='offset points', fontsize=8)
        except:
            pass
        
        ax1.set_xlabel('Volatility (Risk)')
        ax1.set_ylabel('Expected Return')
        ax1.set_title('Efficient Frontier')
        ax1.legend()
        ax1.grid(True, alpha=0.3)
        
        # 2. Portfolio weights comparison
        ax2 = axes[1]
        
        if (max_sharpe_portfolio.get('success', False) and 
            min_vol_portfolio.get('success', False)):
            # Compare top holdings
            sharpe_weights = pd.Series(max_sharpe_portfolio['weights']).sort_values(ascending=False)
            min_vol_weights = pd.Series(min_vol_portfolio['weights']).sort_values(ascending=False)
            
            max_sharpe_top = sharpe_weights.head(8)
            min_vol_top = min_vol_weights.head(8)
            
            # Get common assets
            common_assets = list(set(max_sharpe_top.index) | set(min_vol_top.index))
            
            x_pos = np.arange(len(common_assets))
            width = 0.35
            
            sharpe_values = [max_sharpe_top.get(asset, 0) for asset in common_assets]
            minvol_values = [min_vol_top.get(asset, 0) for asset in common_assets]
            
            ax2.bar(x_pos - width/2, sharpe_values, width, label='Max Sharpe', alpha=0.7)
            ax2.bar(x_pos + width/2, minvol_values, width, label='Min Volatility', alpha=0.7)
            
            ax2.set_xlabel('Assets')
            ax2.set_ylabel('Portfolio Weight')
            ax2.set_title('Portfolio Weights Comparison')
            ax2.set_xticks(x_pos)
            ax2.set_xticklabels(common_assets, rotation=45)
            ax2.legend()
            ax2.grid(True, alpha=0.3)
        else:
            ax2.text(0.5, 0.5, 'Portfolio optimization not successful', 
                    ha='center', va='center', transform=ax2.transAxes)
            ax2.set_title('Portfolio Weights Comparison')
        
        plt.tight_layout()
        plt.show()
    
    @staticmethod
    def plot_risk_analysis(risk_analyzer, max_sharpe_portfolio, min_vol_portfolio, 
                          sharpe_var, sharpe_cvar, minvol_var, minvol_cvar, backtester):
        """Visualize risk analysis results"""
        fig, axes = plt.subplots(2, 2, figsize=(15, 10))
        
        corr_analysis = risk_analyzer.correlation_analysis()
        
        # 1. Correlation heatmap
        ax1 = axes[0, 0]
        corr_matrix = corr_analysis['correlation_matrix']
        if not corr_matrix.empty:
            # Show subset for readability
            subset_corr = corr_matrix.iloc[:15, :15] if len(corr_matrix) > 15 else corr_matrix
            
            sns.heatmap(subset_corr, annot=False, cmap='RdYlBu_r', center=0, 
                       square=True, ax=ax1, cbar_kws={'label': 'Correlation'})
            ax1.set_title('Asset Correlation Matrix')
        else:
            ax1.text(0.5, 0.5, 'Insufficient data for correlation analysis', 
                    ha='center', va='center', transform=ax1.transAxes)
            ax1.set_title('Asset Correlation Matrix')
        
        # 2. VaR comparison
        ax2 = axes[0, 1]
        portfolios = []
        var_values = []
        cvar_values = []
        
        if max_sharpe_portfolio.get('success', False) and sharpe_var is not None:
            portfolios.append('Max Sharpe')
            var_values.append(abs(sharpe_var))
            cvar_values.append(abs(sharpe_cvar) if sharpe_cvar is not None else 0)
        
        if min_vol_portfolio.get('success', False) and minvol_var is not None:
            portfolios.append('Min Volatility')
            var_values.append(abs(minvol_var))
            cvar_values.append(abs(minvol_cvar) if minvol_cvar is not None else 0)
        
        if portfolios:
            x_pos = np.arange(len(portfolios))
            width = 0.35
            
            ax2.bar(x_pos - width/2, var_values, width, label='VaR (5%)', alpha=0.7)
            ax2.bar(x_pos + width/2, cvar_values, width, label='CVaR (5%)', alpha=0.7)
            ax2.set_xlabel('Portfolio')
            ax2.set_ylabel('Risk Measure')
            ax2.set_title('Value at Risk Comparison')
            ax2.set_xticks(x_pos)
            ax2.set_xticklabels(portfolios)
            ax2.legend()
            ax2.grid(True, alpha=0.3)
        else:
            ax2.text(0.5, 0.5, 'No valid portfolios for VaR analysis', 
                    ha='center', va='center', transform=ax2.transAxes)
            ax2.set_title('Value at Risk Comparison')
        
        # 3. Rolling volatility
        ax3 = axes[1, 0]
        if not backtester.returns.empty:
            rolling_vol = backtester.returns.rolling(window=30).std() * np.sqrt(252)
            if not rolling_vol.empty:
                avg_vol = rolling_vol.mean(axis=1)
                avg_vol.plot(ax=ax3, label='Portfolio Volatility', alpha=0.7)
                ax3.set_title('Rolling 30-Day Volatility (Annualized)')
                ax3.set_ylabel('Volatility')
                ax3.legend()
                ax3.grid(True, alpha=0.3)
        else:
            ax3.text(0.5, 0.5, 'No returns data for volatility analysis', 
                    ha='center', va='center', transform=ax3.transAxes)
            ax3.set_title('Rolling 30-Day Volatility')
        
        # 4. Stress test scenarios
        ax4 = axes[1, 1]
        
        # Define stress scenarios
        available_assets = list(backtester.returns.columns)
        stress_scenarios = {
            'Market Crash': {asset: -0.20 for asset in available_assets[:5]},
            'Tech Selloff': {},
            'Financial Crisis': {}
        }
        
        # Populate tech and financial scenarios with available assets
        for asset in available_assets:
            if asset in ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'NVDA', 'TSLA']:
                stress_scenarios['Tech Selloff'][asset] = -0.30
            elif asset in ['JPM', 'BAC', 'WFC', 'GS', 'C']:
                stress_scenarios['Financial Crisis'][asset] = -0.40
        
        stress_results = risk_analyzer.stress_testing(stress_scenarios)
        
        # Calculate portfolio impact for each scenario
        scenario_impacts = []
        scenario_names = []
        
        if max_sharpe_portfolio.get('success', False):
            for scenario_name, scenario_returns in stress_results.items():
                impact = sum(max_sharpe_portfolio['weights'].get(asset, 0) * ret 
                            for asset, ret in scenario_returns.items())
                if abs(impact) > 1e-6:  # Only include meaningful impacts
                    scenario_impacts.append(abs(impact))
                    scenario_names.append(scenario_name)
        
        if scenario_impacts:
            ax4.bar(scenario_names, scenario_impacts, alpha=0.7, color='red')
            ax4.set_title('Stress Test Results (Max Sharpe Portfolio)')
            ax4.set_ylabel('Portfolio Impact')
            ax4.tick_params(axis='x', rotation=45)
            ax4.grid(True, alpha=0.3)
        else:
            ax4.text(0.5, 0.5, 'No stress test results available', 
                    ha='center', va='center', transform=ax4.transAxes)
            ax4.set_title('Stress Test Results')
        
        plt.tight_layout()
        plt.show()


class ReportGenerator:
    """Generate comprehensive backtesting reports"""
    
    @staticmethod
    def save_results(performance_df, max_sharpe_portfolio, min_vol_portfolio, 
                    efficient_frontier, symbols, successful_symbols, period, price_data,
                    sharpe_var=None, sharpe_cvar=None, minvol_var=None, minvol_cvar=None):
        """Save backtesting results and generate comprehensive report"""
        
        # Create output directory
        output_dir = Path('./data/processed/backtesting')
        output_dir.mkdir(parents=True, exist_ok=True)
        
        # Save strategy performance
        if not performance_df.empty:
            performance_df.to_csv(output_dir / 'strategy_performance.csv')
        
        # Save portfolio weights
        if max_sharpe_portfolio.get('success', False):
            pd.Series(max_sharpe_portfolio['weights']).to_csv(output_dir / 'max_sharpe_weights.csv')
        
        if min_vol_portfolio.get('success', False):
            pd.Series(min_vol_portfolio['weights']).to_csv(output_dir / 'min_vol_weights.csv')
        
        # Save efficient frontier
        if not efficient_frontier.empty:
            efficient_frontier.to_csv(output_dir / 'efficient_frontier.csv', index=False)
        
        # Create comprehensive report
        best_strategy_name = ''
        best_sharpe = 0
        best_annual_return = 0
        best_max_drawdown = 0
        
        if not performance_df.empty and 'sharpe_ratio' in performance_df.columns:
            best_idx = performance_df['sharpe_ratio'].idxmax()
            best_strategy_name = str(best_idx)
            best_sharpe = float(performance_df.loc[best_idx, 'sharpe_ratio'])
            best_annual_return = float(performance_df.loc[best_idx, 'annual_return']) if 'annual_return' in performance_df.columns else 0
            best_max_drawdown = float(performance_df.loc[best_idx, 'max_drawdown']) if 'max_drawdown' in performance_df.columns else 0
        
        report = {
            'analysis_date': datetime.now().isoformat(),
            'universe': {
                'total_assets': len(successful_symbols),
                'sectors': list(symbols.keys()),
                'period': period,
                'date_range': {
                    'start': price_data.index.min().isoformat(),
                    'end': price_data.index.max().isoformat()
                }
            },
            'strategies_tested': list(performance_df.index) if not performance_df.empty else [],
            'best_strategy': {
                'name': best_strategy_name,
                'sharpe_ratio': best_sharpe,
                'annual_return': best_annual_return,
                'max_drawdown': best_max_drawdown
            },
            'portfolio_optimization': {
                'max_sharpe': {
                    'success': max_sharpe_portfolio.get('success', False),
                    'expected_return': float(max_sharpe_portfolio.get('expected_return', 0)),
                    'volatility': float(max_sharpe_portfolio.get('volatility', 0)),
                    'sharpe_ratio': float(max_sharpe_portfolio.get('sharpe_ratio', 0))
                },
                'min_volatility': {
                    'success': min_vol_portfolio.get('success', False),
                    'expected_return': float(min_vol_portfolio.get('expected_return', 0)),
                    'volatility': float(min_vol_portfolio.get('volatility', 0)),
                    'sharpe_ratio': float(min_vol_portfolio.get('sharpe_ratio', 0))
                }
            },
            'risk_analysis': {
                'max_sharpe_var': float(sharpe_var) if sharpe_var is not None else None,
                'max_sharpe_cvar': float(sharpe_cvar) if sharpe_cvar is not None else None,
                'min_vol_var': float(minvol_var) if minvol_var is not None else None,
                'min_vol_cvar': float(minvol_cvar) if minvol_cvar is not None else None
            },
            'key_insights': []
        }
        
        # Generate insights
        if best_strategy_name:
            report['key_insights'].append(f"Best performing strategy: {best_strategy_name}")
            report['key_insights'].append(f"Highest Sharpe ratio achieved: {best_sharpe:.3f}")
        
        report['key_insights'].append(f"Portfolio optimization {'successful' if max_sharpe_portfolio.get('success', False) else 'failed'}")
        report['key_insights'].append(f"Risk analysis completed for {len(successful_symbols)} assets")
        
        # Save report
        with open(output_dir / 'backtesting_report.json', 'w') as f:
            json.dump(report, f, indent=2, default=str)
        
        print(f"‚úÖ Backtesting results saved to {output_dir}")
        
        return report
    
    @staticmethod
    def print_summary_report(report, max_sharpe_portfolio, sharpe_var=None):
        """Print summary report to console"""
        print("\nüìã Backtesting Summary Report")
        print("=" * 50)
        print(f"üìä Assets analyzed: {report['universe']['total_assets']}")
        print(f"üìà Strategies tested: {len(report['strategies_tested'])}")
        
        if report['best_strategy']['name']:
            print(f"üèÜ Best strategy: {report['best_strategy']['name']}")
            print(f"üìä Best Sharpe ratio: {report['best_strategy']['sharpe_ratio']:.3f}")
            print(f"üìà Best annual return: {report['best_strategy']['annual_return']:.2%}")
        
        if max_sharpe_portfolio.get('success', False):
            print(f"\nüéØ Optimal Portfolio (Max Sharpe):")
            print(f"  Expected Return: {max_sharpe_portfolio['expected_return']:.2%}")
            print(f"  Volatility: {max_sharpe_portfolio['volatility']:.2%}")
            print(f"  Sharpe Ratio: {max_sharpe_portfolio['sharpe_ratio']:.3f}")
            if sharpe_var is not None:
                print(f"  VaR (5%): {sharpe_var:.3f}")
        
        print("\nüí° Key Recommendations:")
        recommendations = [
            "‚úÖ Diversification across sectors reduces portfolio risk",
            "üìà Technical strategies show varying performance - combine multiple approaches",
            "üéØ Portfolio optimization improves risk-adjusted returns",
            "‚ö†Ô∏è  Regular rebalancing and risk monitoring essential",
            "üìä Backtesting provides valuable insights but past performance doesn't guarantee future results"
        ]
        
        for rec in recommendations:
            print(f"  {rec}")
        
        print("\nüéâ Strategy backtesting pipeline completed successfully!")


class InvestmentBacktestingPipeline:
    """Main pipeline class that orchestrates the entire backtesting process"""
    
    def __init__(self):
        self.data_loader = None
        self.price_data = None
        self.strategies = {}
        self.backtest_results = {}
        self.performance_df = None
        self.max_sharpe_portfolio = None
        self.min_vol_portfolio = None
        self.efficient_frontier = None
        self.risk_analyzer = None
    
    def run_full_pipeline(self, period='3y', initial_capital=100000, commission=0.001):
        """Run the complete backtesting pipeline"""
        
        print("üìä Investment Strategy Backtesting Pipeline Started")
        print("=" * 50)
        
        try:
            # 1. Load Market Data
            print("\nüîÑ Step 1: Loading Market Data")
            self.data_loader = MarketDataLoader()
            self.price_data = self.data_loader.load_data(period)
            
            if self.price_data.empty:
                print("‚ùå No data available. Exiting.")
                return None
            
            print(f"üìä Data loaded successfully: {self.price_data.shape}")
            
            # 2. Generate Technical Strategies
            print("\nüîÑ Step 2: Generating Technical Analysis Signals")
            tech_strategies = TechnicalStrategies(self.price_data)
            
            self.strategies = {
                'MA_Crossover': tech_strategies.moving_average_crossover(20, 50),
                'RSI_Strategy': tech_strategies.rsi_strategy(14, 30, 70),
                'Bollinger_Bands': tech_strategies.bollinger_bands_strategy(20, 2),
                'Momentum': tech_strategies.momentum_strategy(20, 5)
            }
            
            print(f"‚úÖ Generated signals for {len(self.strategies)} strategies")
            for name, signals in self.strategies.items():
                if not signals.empty:
                    buy_signals = (signals == 1).sum().sum()
                    sell_signals = (signals == -1).sum().sum()
                    print(f"  {name}: {buy_signals} buy signals, {sell_signals} sell signals")
            
            # 3. Visualize Strategy Signals
            print("\nüîÑ Step 3: Visualizing Strategy Signals")
            try:
                VisualizationManager.plot_strategy_signals(self.price_data, self.strategies)
            except Exception as e:
                print(f"‚ö†Ô∏è Strategy signal visualization failed: {e}")
            
            # 4. Backtest Strategies
            print("\nüîÑ Step 4: Running Strategy Backtests")
            backtester = StrategyBacktester(self.price_data, initial_capital, commission)
            
            benchmark_returns = None
            if 'SPY' in self.price_data.columns:
                benchmark_returns = self.price_data['SPY'].pct_change().dropna()
            
            for strategy_name, signals in self.strategies.items():
                print(f"Testing {strategy_name}...", end=' ')
                
                try:
                    result = backtester.backtest_strategy(signals, strategy_name)
                    metrics = backtester.calculate_performance_metrics(
                        result['strategy_returns'], benchmark_returns
                    )
                    
                    self.backtest_results[strategy_name] = {
                        'result': result,
                        'metrics': metrics
                    }
                    
                    print(f"‚úÖ Return: {metrics['annual_return']:.2%}, Sharpe: {metrics['sharpe_ratio']:.2f}")
                
                except Exception as e:
                    print(f"‚ùå Error: {e}")
            
            # Create performance summary
            if self.backtest_results:
                self.performance_df = pd.DataFrame({
                    name: results['metrics'] 
                    for name, results in self.backtest_results.items()
                }).T
                
                print("\nüìä Strategy Performance Summary:")
                print(self.performance_df.round(4))
            else:
                print("‚ùå No successful backtests")
                self.performance_df = pd.DataFrame()
            
            # 5. Visualize Strategy Performance
            print("\nüîÑ Step 5: Visualizing Strategy Performance")
            try:
                VisualizationManager.plot_strategy_performance(self.backtest_results, benchmark_returns)
            except Exception as e:
                print(f"‚ö†Ô∏è Strategy performance visualization failed: {e}")
            
            # 6. Portfolio Optimization
            print("\nüîÑ Step 6: Portfolio Optimization")
            try:
                optimizer = PortfolioOptimizer(backtester.returns)
                
                # Find optimal portfolios
                print("Finding optimal portfolios...")
                
                # Maximum Sharpe ratio portfolio
                self.max_sharpe_portfolio = optimizer.optimize_portfolio('sharpe')
                if self.max_sharpe_portfolio.get('success', False):
                    print(f"‚úÖ Max Sharpe portfolio: {self.max_sharpe_portfolio.get('sharpe_ratio', 0):.3f}")
                else:
                    print("‚ùå Max Sharpe optimization failed")
                
                # Minimum volatility portfolio
                self.min_vol_portfolio = optimizer.optimize_portfolio('volatility')
                if self.min_vol_portfolio.get('success', False):
                    print(f"‚úÖ Min volatility portfolio: {self.min_vol_portfolio.get('volatility', 0):.3f}")
                else:
                    print("‚ùå Min volatility optimization failed")
                
                # Generate efficient frontier
                print("Generating efficient frontier...")
                self.efficient_frontier = optimizer.efficient_frontier(30)
                
            except Exception as e:
                print(f"‚ùå Portfolio optimization failed: {e}")
                self.max_sharpe_portfolio = {'success': False}
                self.min_vol_portfolio = {'success': False}
                self.efficient_frontier = pd.DataFrame()
            
            # Display optimal portfolios
            if self.max_sharpe_portfolio.get('success', False):
                print("\nüìä Maximum Sharpe Ratio Portfolio:")
                sharpe_weights = pd.Series(self.max_sharpe_portfolio['weights']).sort_values(ascending=False)
                print(f"Expected Return: {self.max_sharpe_portfolio['expected_return']:.2%}")
                print(f"Volatility: {self.max_sharpe_portfolio['volatility']:.2%}")
                print(f"Sharpe Ratio: {self.max_sharpe_portfolio['sharpe_ratio']:.3f}")
                print("\nTop 10 Holdings:")
                print(sharpe_weights.head(10).apply(lambda x: f"{x:.1%}"))
            
            if self.min_vol_portfolio.get('success', False):
                print("\nüìä Minimum Volatility Portfolio:")
                min_vol_weights = pd.Series(self.min_vol_portfolio['weights']).sort_values(ascending=False)
                print(f"Expected Return: {self.min_vol_portfolio['expected_return']:.2%}")
                print(f"Volatility: {self.min_vol_portfolio['volatility']:.2%}")
                print(f"Sharpe Ratio: {self.min_vol_portfolio['sharpe_ratio']:.3f}")
                print("\nTop 10 Holdings:")
                print(min_vol_weights.head(10).apply(lambda x: f"{x:.1%}"))
            
            # 7. Visualize Portfolio Optimization
            print("\nüîÑ Step 7: Visualizing Portfolio Optimization")
            try:
                VisualizationManager.plot_portfolio_optimization(
                    optimizer, self.max_sharpe_portfolio, self.min_vol_portfolio, self.efficient_frontier
                )
            except Exception as e:
                print(f"‚ö†Ô∏è Portfolio optimization visualization failed: {e}")
            
            # 8. Risk Analysis
            print("\nüîÑ Step 8: Performing Risk Analysis")
            try:
                self.risk_analyzer = RiskAnalyzer(backtester.returns)
                
                # Calculate risk metrics for optimal portfolios
                sharpe_var = sharpe_cvar = minvol_var = minvol_cvar = None
                
                if self.max_sharpe_portfolio.get('success', False):
                    sharpe_weights_dict = self.max_sharpe_portfolio['weights']
                    
                    sharpe_var = self.risk_analyzer.calculate_var(0.05, sharpe_weights_dict)
                    sharpe_cvar = self.risk_analyzer.calculate_cvar(0.05, sharpe_weights_dict)
                    
                    print(f"üìä Max Sharpe Portfolio Risk Metrics:")
                    print(f"  VaR (5%): {sharpe_var:.3f}")
                    print(f"  CVaR (5%): {sharpe_cvar:.3f}")
                
                if self.min_vol_portfolio.get('success', False):
                    minvol_weights_dict = self.min_vol_portfolio['weights']
                    
                    minvol_var = self.risk_analyzer.calculate_var(0.05, minvol_weights_dict)
                    minvol_cvar = self.risk_analyzer.calculate_cvar(0.05, minvol_weights_dict)
                    
                    print(f"\nüìä Min Volatility Portfolio Risk Metrics:")
                    print(f"  VaR (5%): {minvol_var:.3f}")
                    print(f"  CVaR (5%): {minvol_cvar:.3f}")
                
                # Correlation analysis
                corr_analysis = self.risk_analyzer.correlation_analysis()
                
                if not corr_analysis['highest_correlations'].empty:
                    print("\nüîó Correlation Analysis:")
                    print("\nHighest Correlations:")
                    print(corr_analysis['highest_correlations'].head())
                    
                    print("\nLowest Correlations:")
                    print(corr_analysis['lowest_correlations'].head())
                
            except Exception as e:
                print(f"‚ùå Risk analysis failed: {e}")
                sharpe_var = sharpe_cvar = minvol_var = minvol_cvar = None
            
            # 9. Visualize Risk Analysis
            print("\nüîÑ Step 9: Visualizing Risk Analysis")
            try:
                VisualizationManager.plot_risk_analysis(
                    self.risk_analyzer, self.max_sharpe_portfolio, self.min_vol_portfolio,
                    sharpe_var, sharpe_cvar, minvol_var, minvol_cvar, backtester
                )
            except Exception as e:
                print(f"‚ö†Ô∏è Risk analysis visualization failed: {e}")
            
            # 10. Generate and Save Report
            print("\nüîÑ Step 10: Generating Report")
            try:
                report = ReportGenerator.save_results(
                    self.performance_df, self.max_sharpe_portfolio, self.min_vol_portfolio,
                    self.efficient_frontier, self.data_loader.symbols, self.data_loader.successful_symbols,
                    period, self.price_data, sharpe_var, sharpe_cvar, minvol_var, minvol_cvar
                )
                
                ReportGenerator.print_summary_report(report, self.max_sharpe_portfolio, sharpe_var)
            except Exception as e:
                print(f"‚ö†Ô∏è Report generation failed: {e}")
                report = {}
            
            return {
                'data_loader': self.data_loader,
                'price_data': self.price_data,
                'strategies': self.strategies,
                'backtest_results': self.backtest_results,
                'performance_df': self.performance_df,
                'max_sharpe_portfolio': self.max_sharpe_portfolio,
                'min_vol_portfolio': self.min_vol_portfolio,
                'efficient_frontier': self.efficient_frontier,
                'risk_analyzer': self.risk_analyzer,
                'report': report
            }
            
        except Exception as e:
            print(f"‚ùå Pipeline failed with error: {e}")
            import traceback
            traceback.print_exc()
            return None


def main():
    """Main function to run the investment backtesting pipeline"""
    
    try:
        # Initialize and run the pipeline
        pipeline = InvestmentBacktestingPipeline()
        results = pipeline.run_full_pipeline(
            period='3y',           # Analysis period
            initial_capital=100000, # Starting capital
            commission=0.001       # Transaction cost (0.1%)
        )
        
        if results:
            print("\n‚úÖ Pipeline completed successfully!")
        else:
            print("\n‚ùå Pipeline failed to complete")
        
        # Optional: Return results for further analysis
        return results
        
    except KeyboardInterrupt:
        print("\n‚ùå Pipeline interrupted by user")
    except Exception as e:
        print(f"\n‚ùå Pipeline failed with error: {e}")
        import traceback
        traceback.print_exc()


if __name__ == "__main__":
    # Run the main pipeline
    results = main()
    
    # Optional: Access specific results
    if results:
        print(f"\nüìä Final Results Available:")
        print(f"  - Price data shape: {results['price_data'].shape}")
        print(f"  - Strategies tested: {len(results['strategies'])}")
        
        if not results['performance_df'].empty:
            best_strategy = results['performance_df'].sort_values('sharpe_ratio', ascending=False).index[0]
            print(f"  - Best strategy: {best_strategy}")
        
        max_sharpe_success = results['max_sharpe_portfolio'].get('success', False)
        print(f"  - Portfolio optimization: {'‚úÖ Success' if max_sharpe_success else '‚ùå Failed'}")
        
        # Example of accessing specific results:
        # best_strategy = results['performance_df'].sort_values('sharpe_ratio', ascending=False).iloc[0]
        # optimal_portfolio = results['max_sharpe_portfolio']
        # correlation_matrix = results['risk_analyzer'].correlation_analysis()['correlation_matrix']
                
```
